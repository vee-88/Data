SETTING QUARTERS


```python
def get_quarter(month):
    if month in ['July', 'August', 'September']:
        return 1
    elif month in ['October', 'November', 'December']:
        return 2
    elif month in ['January', 'February', 'March']:
        return 3
    else:
        return 4

# Assuming your DataFrame has a 'Month' column representing the month names
df_new['Quarter'] = df_new['Month'].apply(get_quarter)
```

```python
df_new.isnull().sum()
```

```python
df_new.head(2)
```

```python
df_new['Quarter'] = df_new['Quarter'].replace(1, 'Quarter 1')
df_new['Quarter'] = df_new['Quarter'].replace(2, 'Quarter 2')
df_new['Quarter'] = df_new['Quarter'].replace(3, 'Quarter 3')
df_new['Quarter'] = df_new['Quarter'].replace(4, 'Quarter 4')
```

REVENUE PER QUARTER

```python
quarter_revenue = df_new.groupby('Quarter')['Amount Paid'].sum()
quarter_revenue_formatted = quarter_revenue.apply(lambda x: '{:,.0f}'.format(x))
quarter_revenue_formatted
```

```python
import matplotlib.pyplot as plt
quarter_revenue = df_new.groupby('Quarter')['Amount Paid'].sum()

quarters = quarter_revenue.index
total_amounts = quarter_revenue.values

plt.pie(total_amounts, labels=None, autopct='%1.1f%%')
plt.title('Total Revenue Per Department')
plt.axis('equal') 
plt.legend(quarters, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
